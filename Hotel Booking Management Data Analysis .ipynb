```python
# Create new columns
df['booking_date'] = df['reservation_status_date'].dt.date  # Full date
df['booking_month'] = df['reservation_status_date'].dt.month  # Month

df['room_type'] = df['assigned_room_type']  # Actual room type

df['total_price'] = df['adr'] * (df['stays_in_weekend_nights'] + df['stays_in_week_nights'])  # Revenue calculation
```

```python
# Save the updated DataFrame to a new file
df.to_csv("hotel_bookings_updated.csv", index=False)
```

```python
# Data Preprocessing
df.dropna(inplace=True)  # Removing missing values
df['booking_date'] = pd.to_datetime(df['booking_date'])
```

```python
# Feature Engineering
df['booking_month'] = df['booking_date'].dt.month
```

```python
df = df[df['adr']<5000]
```

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
```

```python
# Visualization 1 : Reservation status in different hotels

plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
ax1.bar_label(ax1.containers[0])
ax1.bar_label(ax1.containers[1])
plt.title('Reservation status in different hotels', size = 10)
plt.xlabel('Hotels')
plt.ylabel('Number of Reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Visualization 2 : Average Daily Rate in City And Resort Hotel
plt.figure(figsize = (15,8))
plt.title('Average Daily Rate in City And Resort Hotel', fontsize = 10)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 15)
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
# Calculate average daily rate for cancelled reservations
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Calculate average daily rate for not cancelled reservations
# Assuming reservations_data is your main DataFrame containing all reservations
not_cancelled_data = df[df['reservation_status'] != 'Cancelled'] 

not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)
```

```python
# Visualization 3 : Average Daily Rate in Reservation_status_date
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled', color='blue')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled', color='red')
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid()
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Visualization 4 : Monthly Booking & Cancellation Trends
plt.figure(figsize=(12,6))
sns.lineplot(x=df['booking_month'], y=df['is_canceled'], estimator=sum, ci=None)
plt.title("Monthly Booking & Cancellation Trends")
plt.xlabel("Month")
plt.ylabel("Cancellations")
plt.show()
```

```python
# Visualization 5: Total Revenue by Hotel Type
hotel_revenue = df.groupby('hotel')['total_price'].sum().reset_index()
plt.figure(figsize=(8, 8))
ax = sns.barplot(x='hotel', y='total_price', data=hotel_revenue, palette='viridis')

# Adding value annotations on top of the bars
ax.bar_label(ax.containers[0]) # Format to 2 decimal places
ax.bar_label(ax.containers[1]) 

plt.title("Total Revenue by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Total Revenue")
plt.xticks(rotation=40)  # Rotate x labels for better readability if needed
plt.show()
```

```python
# Visualization 6: Guest Segmentation by Customer Type
plt.figure(figsize=(8,5))
ax = sns.countplot(x='customer_type', data=df, palette='coolwarm')

# Adding value annotations on top of the bars
for container in ax.containers:
    ax.bar_label(container)
    
plt.title("Guest Segmentation by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 7: Weekend vs. Weekday Stays
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
night_type = df[['stays_in_weekend_nights', 'stays_in_week_nights']].sum()
plt.figure(figsize=(6,4))
plt.pie(night_type, labels=['Weekend Nights', 'Weekday Nights'], autopct='%1.1f%%', colors=['#FF9999', '#66B3FF'])
plt.title("Comparison of Weekend vs. Weekday Stays")
plt.show()
```

```python
# Visualization 8: Number of Bookings Per Month
plt.figure(figsize=(12,6))
ax = sns.countplot(data=df, x='arrival_date_month', order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
# Adding value annotations on top of the bars
for container in ax.containers:
    ax.bar_label(container)
plt.title("Number of Bookings Per Month")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 9: Booking Cancellation Rate 
plt.figure(figsize=(8,5))
ax = sns.countplot(data=df, x='is_canceled', palette='coolwarm')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])
plt.title("Booking Cancellation Rate")
plt.show()
```

```python
# Visualization 10: Average Daily Rate (ADR) Per Month
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='arrival_date_month', y='adr', hue='hotel', estimator=np.mean)
plt.title("Average Daily Rate (ADR) Per Month")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 11: Top 10 Countries by Bookings
plt.figure(figsize=(12,6))
country_counts = df['country'].value_counts().head(10)
ax = sns.barplot(x=country_counts.index, y=country_counts.values, palette='viridis')

# Adding value annotations on top of the bars
for container in ax.containers:
    ax.bar_label(container)

plt.title("Top 10 Countries by Bookings")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 12: Market Segment Distribution
plt.figure(figsize=(12,6))
ax = sns.countplot(data=df, y='market_segment', order=df['market_segment'].value_counts().index, palette='pastel')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[3])
ax.bar_label(ax.containers[4])
ax.bar_label(ax.containers[5])
plt.title("Market Segment Distribution")
plt.show()
```
